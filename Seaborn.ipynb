```
# Seaborn contains lots of statistical tools which helps us in exploratory data analysis
import seaborn as sns
```

```
df = sns.load_dataset("tips")
df.dtypes
# tip is our dependent feature.
# total_bill, gender, smoker, day, time.. all these are independent features.
```

```
df.head()
```

**Correlation with Heatmap**

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection



```
# Correlation helps us to find interrelationship.

df.corr()

# Object type features i.e., categorial features can't have a correlation only value(int, float) type features can have a relation.
```

```
 sns.heatmap(df.corr())
```

### JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot



**Univariate Analysis** : 1 independent variable

```
#sns.joinplot(df)
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
# Draws probablity density function
```

### PairPlot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables



**Bi-variate** : 2 independent variable.

```
sns.pairplot(df)
# It will do permutation and combination of all the non-categorial features.
```

```
sns.pairplot(df,hue='sex')
```

```
df['sex'].value_counts()
```

```
df['smoker'].value_counts()
```

### DistPlot

Dist plot helps us to check the distribution of the columns feature



```
sns.distplot(df['tip'])
```

```
sns.distplot(df['tip'],kde = False,  bins = 10 )
```

### Categorial Plots

Seaborn also helps us in doing the analysis on Categorial Data Points. In this section we will  discuss about : 
*   boxplot
*   violinplot
*   countplot
*   bar plot



```
# Count Plot

sns.countplot('sex', data=df)
```

```
sns.countplot('smoker',data=df)
```

```
sns.countplot(y='sex',data=df)
```

```
sns.countplot('day',data=df)
```

```
# Bar Plot

sns.barplot(x='sex', y='total_bill',data=df)
```

```
df.head()
```

**Box Plot** 

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summmary.


```
sns.boxplot(x='sex',y = 'total_bill',data=df)
```

```
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```
sns.boxplot(data = df, orient = 'v')
```

```
# categorize my data based on some other categories

sns.boxplot(y="total_bill", x="day", hue="smoker",data=df)

```

**Violin**

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot



```
# it is combination of kde annd boxplot
sns.violinplot(y="total_bill", x="day", data=df,palette='rainbow')
```
