### **Matplotlib**
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

*   Generally easy to get started for simple plots   
*   Support for custom labels and texts
*   Great control of every element in a figure
*   High-quality output in many formats
*   Very customizable in general


```
import matplotlib.pyplot as plt

%matplotlib inline   

# you need not write plot.show() if you write above statement.
```

```
import numpy as np
```

```
# Simple Examples:

x = np.arange(0,10)
y = np.arange(11,21)
```

**Ploting using matplotlib**

```
# plt scatter

plt.scatter(x,y,c='g') # plt.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=matplotlib.cbook.deprecation._deprecated_parameter_class instance, edgecolors=None, *, plotnonfinite=False, data=None, **kwargs)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```
y = x*x
```

```
# plt plot

plt.plot(x,y,'r*', linestyle='dashed', linewidth=2, markersize=12)    # Basically gives styling options.
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

**Creating** **subplots**

```
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*-')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

```
x = np.arange(1,11)
y = 3*x + 5
plt.title("Matplotlib demo")
plt.xlabel("X-axis caption")
plt.ylabel("Y-axis caption")
plt.plot(x,y)
```

```
# Computing x,y co-ordinate for points on a sine curve

x = np.arange(0, 4*np.pi,  0.001)
y = np.sin(x)
plt.title("Sine wave form")

# Plot the points using matplotlib
plt.plot(x,y,'g')
```

```
# Subplots
plt.subplot(2,1,1)
plt.plot(x,y,'g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine wave')



plt.subplot(2,1,2)
plt.plot(y,x,'b')
plt.xlabel('Y-axis')
plt.ylabel('X-axis')
plt.title('Cosec wave')
```

**Bar plot**

```
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

**Histogram**

```
a = np.array([22,87,5,43,73,55,54,56,57,11,20,23,51,5,79,78,31,27])
plt.hist(a)
plt.title("Histogram")
plt.xlabel("Numbers")
plt.ylabel("Dens")
```

**Box plot using matplotlib**

```
data = [np.random.normal(0,std,100) for std in range(1,4)]

# Rectangular Box plot (used for getting percentile) 
plt.boxplot(data,vert=True,patch_artist=True)
```

**Pie chart**

```
# Data  to plot

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
```
